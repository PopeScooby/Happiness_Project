```python
# Import library
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] =14
plt.rcParams['font.weight']= 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
# Import dataset
# The happiness data files 2021 and trending
happiness_t_file= "world-happiness-report.csv"
happiness_file = "world-happiness-report-2021.csv"
```

```python
happiness_t_df = pd.read_csv(happiness_t_file, index_col=0)
happiness_t_df.head(20)
```

```python
happiness_df = pd.read_csv(happiness_file, index_col=0)
happiness_df.head(20)
```

```python
sns.lmplot(x='Ladder score',y='Logged GDP per capita',data=happiness_df,aspect=2,height=6)
plt.xlabel('Ladder Score: as Independent variable')
plt.ylabel('Logged GDP per capita: as Dependent variable')
plt.title('Ladder Score Vs GDP per capita');
```

```python
sns.scatterplot(x='Logged GDP per capita',y='Ladder score',data=happiness_df,palette='magma',hue='Regional indicator', legend = False)
plt.xlabel('Logged GDP per capita: as Independent variable')
plt.ylabel('Ladder Score: as Dependent variable')
plt.title('Ladder Score vs GDP per capita by Region');
```

```python
happiness_df.describe()
```

```python
# correlation plot
corr = happiness_df.corr()
sns.heatmap(corr,  cmap = 'Wistia', annot= False);
```

```python
happiness_df.groupby('Regional indicator').agg(['mean','min','max'])['Ladder score']
```

```python
f = plt.figure(figsize=(14,6))
ax = f.add_subplot(121)
sns.scatterplot(x='Logged GDP per capita',y='Ladder score',data=happiness_df,palette='magma',hue='Regional indicator', legend = False)
ax.set_title('Scatter plot of GDP per capita vs Ladder Score')

ax = f.add_subplot(122)
sns.scatterplot(x='Healthy life expectancy',y='Ladder score',data=happiness_df,palette='viridis',hue='Regional indicator',legend = False)
ax.set_title('Healthy life expectancy vs Ladder Score')
#plt.savefig('sc.png');
```
