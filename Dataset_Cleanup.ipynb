```python
# Import library
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] =14
plt.rcParams['font.weight']= 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
# Import dataset
# The happiness data files 2021 and trending
happiness_t_file= "Data/world-happiness-report.csv" #checked
happiness_file = "Data/world-happiness-report-2021.csv" #checked
covid_data = "Data/kaggle-covid/worldometer_data_test.csv" #checked
screen_time = "Data/Clean_Average_Screen_Time_per_Country_2021.csv" #checked
population_density = "Data/Country_Population_Land_Area_Density.csv" #checked
alcohol_consumption = "Data/alcohol_consumption_bycountry.csv" #checked
meat_consumption = "Data/Meat_consumption.csv" #checked
median_age = "Data/Median_age.csv" #checked
suicide_rate = "Data/Suicide_rate.csv" #checked
```

```python
# Convert datasets to dataframes
happiness_t_file_df = pd.read_csv(happiness_t_file)
happiness_file_2021_df = pd.read_csv(happiness_file)
covid_data_df = pd.read_csv(covid_data)
screen_time_df = pd.read_csv(screen_time)
population_density_df = pd.read_csv(population_density)
alcohol_consumption_df = pd.read_csv(alcohol_consumption)
meat_consumption_df = pd.read_csv(meat_consumption)
median_age_df = pd.read_csv(median_age)
suicide_rate_df = pd.read_csv(suicide_rate)
```

```python
happiness_t_file_df.head()
```

```python
#rename columns in happiness_t_file dataframe
happiness_t_file_df.rename(columns = {'Country name': 'Country_Name'}, inplace = True)
happiness_t_file_df.rename(columns = {'Regional indicator': 'Regional_Indicator'}, inplace = True)
happiness_t_file_df.rename(columns = {'Life Ladder': 'Ladder_Score'}, inplace = True)
happiness_t_file_df.rename(columns = {'Log GDP per capita': 'Logged_GDP_per_Capita'}, inplace = True)
happiness_t_file_df.rename(columns = {'Healthy life expectancy at birth': 'Healthy_Life_Expectancy'}, inplace = True)
happiness_t_file_df.rename(columns = {'Freedom to make life choices': 'Freedom_to_make_life_choices'}, inplace = True)
happiness_t_file_df.rename(columns = {'Perceptions of corruption': 'Perceptions_of_corruption'}, inplace = True)
happiness_t_file_df.rename(columns = {'Positive affect': 'Positive_Affect'}, inplace = True)
happiness_t_file_df.rename(columns = {'Negative affect': 'Negative_affect'}, inplace = True)
```
