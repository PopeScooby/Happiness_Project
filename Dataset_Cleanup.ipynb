```python
# Import dataset
# The happiness data files 2021 and trending
happiness_t_file= "Data/world-happiness-report.csv" #check
happiness_file = "Data/world-happiness-report-2021.csv" #check
covid_data = "Data/kaggle-covid/worldometer_data_test.csv"
screen_time = "Data/Clean_Average_Screen_Time_per_Country_2021.csv"
population_density = "Data/Country_Population_Land_Area_Density.csv"
alcohol_consumption = "Data/alcohol_consumption_bycountry.csv" #check
meat_consumption = "Data/Meat_consumption.csv"
median_age = "Data/Median_age.csv"
suicide_rate = "Data/Suicide_rate.csv"
```

```python
# Convert datasets to dataframes
happiness_t_file_df = pd.read_csv(happiness_t_file)
happiness_file_2021_df = pd.read_csv(happiness_file)
covid_data_df = pd.read_csv(covid_data)
screen_time_df = pd.read_csv(screen_time, index_col=0)
population_density_df = pd.read_csv(population_density, index_col=0)
alcohol_consumption_df = pd.read_csv(alcohol_consumption)
meat_consumption_df = pd.read_csv(meat_consumption, index_col=0)
median_age_df = pd.read_csv(median_age, index_col=0)
suicide_rate_df = pd.read_csv(suicide_rate, index_col=0)
```
